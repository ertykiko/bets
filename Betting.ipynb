```python
web = "https://www.betclic.pt/futebol-s1"
ser = Service("C:/Users/Francisco/Documents/bets/chromedriver.exe")
op = webdriver.ChromeOptions()

driver = webdriver.Chrome(service=ser, options=op)

driver.get(web)

Xpath = '/html/body/app-desktop/div[1]/div/bcdk-content-scroller/div/sports-all-offer/sports-events-list/bcdk-vertical-scroller/div/div[2]/div/div/div[2]'

all_events = driver.find_element(By.XPATH, Xpath)
```
